# 데이터 불러오기 및 EDA
train : 2022_09~2023_08
test : 2023_09, 2023_10, 2023_11

```python
import pandas as pd
import numpy as np

train_files = ['2022_09.csv', '2022_10.csv', '2022_11.csv', '2022_12.csv', '2023_01.csv', '2023_02.csv',
        '2023_03.csv', '2023_04.csv', '2023_05.csv', '2023_06.csv', '2023_07.csv', '2023_08.csv']

train_df = [pd.read_csv(f) for f in train_files]

train = pd.concat(train_df)
```

```python
#train 데이터 살펴보기
train.head()
```

```python
train.info()
```

결측치가 없음을 확인할 수 있었다.

## 퇴근 및 출근 시간대 선정

```python
#편의를 위해 00시를 0시로 바꾼다.
train = train.rename(columns={'00시승차총승객수': '0시승차총승객수'})
train = train.rename(columns={'00시하차총승객수': '0시하차총승객수'})
```

```python
train.head(3)
```

```python
#출근시간, 퇴근시간을 정하기 위해 먼저 데이터프레임에서 승차총승객수가 가장 높은 3개의 시간대를 알아 본다.

time_columns = [f'{hour}시승차총승객수' for hour in range(24)]

sum_ride = train[time_columns].sum()

# 합이 가장 큰 상위 5개의 칼럼을 찾는다.

top_5_columns = sum_ride.nlargest(5).index.tolist()

print(top_5_columns)
```

따라서, 7,8,9시를 출근시간, 16,17,18시를 퇴근시간으로 정한다.

필요없는 칼럼을 삭제한다.

```python
# '등록일자' 칼럼 삭제
train = train.drop('등록일자', axis=1)

train.head()
```

## 카테고리 3개로 나누기
### train에서 버스타입명 카테고리 칼럼을 추가한다. 

###### 0: 서울 시내버스 1: 서울 마을버스 2:서울 광역버스

```python
#'버스타입명' 칼럼을 생성하고 각각 0, 1, 2 값을 할당하는 함수를 정의한다.
def assign_bus_type(category):
    if category in ['서울마을버스']:
        return 1
    elif category in ['서울광역버스']:
        return 2
    else:
        return 0

#'교통수단타입명' 칼럼을 사용하여 '버스타입명' 칼럼을 생성한다.
train['버스타입명'] = train['교통수단타입명'].apply(assign_bus_type)

train.head()
```

```python
#버스타입명을 기준으로 데이터 정렬함
train_sorted = train.sort_values(by=['버스타입명'])
train_sorted.head()
```

```python
#칼럼명을 굳이 바꿔야 하나 의문이 들긴 한다. 하지만 일단 코드는 작성해두었다.

train = train.rename(columns={'사용년월': 'month'}, )
train = train.rename(columns={'표준버스정류장ID': 'bus_route_id'})
train = train.rename(columns={'노선번호': 'bus_id'})
train = train.rename(columns={'노선명': 'route_name'})
train = train.rename(columns={'역명': 'station_name'})
train = train.rename(columns={'버스타입명': 'bus_type_name'})
rename_dict = {f'{hour}시승차총승객수': f'{hour} ride' for hour in range(24)}
rename_dict.update({f'{hour}시하차총승객수': f'{hour} take-off' for hour in range(24)})
train = train.rename(columns=rename_dict)
#사용하지 않을 것 같은 칼럼 제외
train = train.drop('버스정류장ARS번호', axis=1)
train = train.drop('교통수단타입코드', axis=1)
train = train.drop('교통수단타입명', axis=1)
train.head()
```
